```python
import pickle
from matplotlib import pyplot as plt
import numpy as np
```

```python
results_switch, results_hashmap = None, None

with open("results_switch.pkl", "rb") as file:
    results_switch = pickle.load(file)

with open("results_hashmap.pkl", "rb") as file:
    results_hashmap = pickle.load(file)

```

```python
sizes = [1, 2, 4, 8, 16, 32, 64, 128]
switch_mean = [np.mean(results_switch[size]) for size in sizes]
switch_std = [np.std(results_switch[size]) for size in sizes]

hash_mean = [np.mean(results_hashmap[size]) for size in sizes]
hash_std = [np.std(results_hashmap[size]) for size in sizes]

plt.figure(figsize=(10, 6))

# Plotting results for the first dictionary
plt.errorbar(sizes, switch_mean, yerr=switch_std, fmt='o-', label="Switch", capsize=5)

# Plotting results for the second dictionary
plt.errorbar(sizes, hash_mean, yerr=hash_std, fmt='s-', label="Hash", capsize=5)

plt.xlabel("Size")
plt.ylabel("Mean Value")
plt.title("Comparison of Results")
plt.legend()
plt.grid(True)
```
